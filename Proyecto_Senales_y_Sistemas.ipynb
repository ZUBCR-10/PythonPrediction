```python
#Grafica de la muestra de audio.
def visualize():
   
    # Abrimos el audio.
    raw = wave.open('/content/piano-C4.wav')
     
    # Lee todos los frames
    # -1 indica el todos o el maximo de frames
    signal = raw.readframes(-1)
    signal = np.frombuffer(signal, dtype ="int16")
     
    # Obtiene el frame rate
    f_rate = raw.getframerate()
 
    # para trazar el eje (x) en segundos, necesita obtener la velocidad de
    # fotogramas y dividir por el tamaño de su señal para crear un vector de
    # tiempo espaciado.
    time = np.linspace(
        0, # start
        len(signal) / f_rate,
        num = len(signal)
    )
 
    # Use la función matplotlib.figure() para trazar el gráfico.
    # Crea un grafico.
    plt.figure(1)
     
    # Titulo del Grafico.
    plt.title("Sound Wave")
     
    # Se asigna el eje x.
    plt.xlabel("Time")
    
    # Ecuacion que grafica. 
    plt.plot(time, signal)
     
    # Se imprime en pantalla la grafica.
    plt.show()
 
 
if __name__ == "__main__":
   
    # obtiene el valor de la línea de comando
    path = sys.argv[1]
    visualize()
 
    
```

La grafica anterior representa la forma de onda de la muestra de audio, de manera que podemos observar el comportamiento de la frecuencia-tiempo, la grafica va a cambiar dependiendo de la muestra de audio que usemos.

En esta parte se requiere cargar los archivos: mylib.py y piano-C4.wav, para simplemnete ejecuatar el codigo sin tener que hacer modificaciones. 

Importamos las bibliotecas.

```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
plt.style.use('seaborn-dark')
```

Cambiamos la frecuencia de la nota musical en: frequency = note_freqs['C4'] por la que corresponda a nuestra muestra de audio.

Podemos cambiar la muestra de audio en:  sample_rate, middle_c = wavfile.read('muestra_de_audio.wav')

Tambien cambiamos los nombres de los archivos que se van a crear, en este caso solo se cambian las lineas en que se repite piano-C4.wav (pure, sound_wave, Spectrum of Middle G Recording, Spectrum, synthetic) en estos campos se cambian por la nueva muestra de audio.

Ejecutamos el codigo.

```python
import mylib
# Obtener frecuencia media 'C'
note_freqs = mylib.get_piano_notes()
frequency = note_freqs['C4']

# Onda sinusoidal pura
sine_wave = mylib.get_sine_wave(frequency, duration=2, amplitude=2048)
wavfile.write('purepiano-C4.wav', rate=44100, data=sine_wave.astype(np.int16))

# Abrimos el archivo .wav
sample_rate, middle_c = wavfile.read('piano-C4.wav')

# Trazar onda de sonido
plt.plot(middle_c[500:2500])
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sound Wave of Middle G on piano-C4')
plt.grid()
plt.savefig('piano-C4-sound_wave.jpg')
plt.show()



# Aplicaremos la transformada rápida de Fourier
t = np.arange(middle_c.shape[0])
freq = np.fft.fftfreq(t.shape[-1])*sample_rate
sp = np.fft.fft(middle_c) 



# Trazar espectro de magnitud (Amplitud/Frecuencia)
plt.plot(freq, abs(sp.real))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Spectrum of Middle G Recording on piano-C4')
plt.xlim((0, 2000))
plt.grid()
plt.savefig('spectrumpiano-C4.jpg')
plt.show()

# Obtener frecuencia positiva
idx = np.where(freq > 0)[0]
freq = freq[idx]
sp = sp[idx]

# Obtener frecuencias dominantes
sort = np.argsort(-abs(sp.real))[:100]
dom_freq = freq[sort]

# Redondear y calcular la relación de amplitud
freq_ratio = np.round(dom_freq/frequency)
unique_freq_ratio = np.unique(freq_ratio)
amp_ratio = abs(sp.real[sort]/np.sum(sp.real[sort]))
factor = np.zeros((int(unique_freq_ratio[-1]), ))
for i in range(factor.shape[0]):
    idx = np.where(freq_ratio==i+1)[0]
    factor[i] = np.sum(amp_ratio[idx])
factor = factor/np.sum(factor)

# Construir series armónicas
note = mylib.apply_overtones(frequency, duration=2.5, factor=factor)

# Aplicar pesos suaves ADSR (modelo para describir el comportamiendo del sonido).
weights = mylib.get_adsr_weights(frequency, duration=2.5, length=[0.05, 0.25, 0.55, 0.15],
                           decay=[0.075,0.02,0.005,0.1], sustain_level=0.1)

# Guardamos el archivo.
data = note*weights
data = data*(4096/np.max(data)) # Adjusting the Amplitude 
wavfile.write('syntheticpiano-C4.wav', sample_rate, data.astype(np.int16))
```

Finalmente logramos obtener la forma de onda de la señal y el espectro de magnitud de la serie de Fourier.

El espectro de la senal simplemente traza la onda de sonido, permitiendonos ver mas facilmente las caracteristicas de esta seañal.

El proceso que separa señales de diferentes frecuencias de una señal de entrada es mediante la transformada de Fourier (espectro de frecuencia discreta), esta operación nos permite clasificar las ondas sonoras por rango de frecuencias.
